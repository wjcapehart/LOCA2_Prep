```python
root_url = "/projects/ECEP/LOCA_MACA_Ensembles/LOCA2/LOCA2_CONUS/Original_CONUS/"

inventory_file = "./LOCA2_inventory_min_errors_short.csv"

inventory = pd.read_csv(filepath_or_buffer    = inventory_file)



display(inventory)
```

```python
precmap = colortables.get_colortable("precipitation")
tmxcmap = plt.cm.get_cmap('autumn').reversed()

tmncmap = plt.cm.get_cmap('winter')

clevs = [0, 1, 2.5, 5, 7.5, 10, 15, 20, 30, 40,
         50, 70, 100, 150, 200, 250, 300, 400, 500, 600, 750]


tind = 31410 #- 182

figsize0 = [ 8,  5]

for index, inventory_row in inventory.iterrows():
    
    print(index, inventory_row.Model, inventory_row.Member)
    
    # tasmax 
    var  = "tasmax"
    cmap = tmxcmap
    
    fig = plt.figure(figsize = figsize0)
    
    scen = 0
    if (inventory_row[scen+1]):
        scenario = "historical"

        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        

        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            inarr0x = xf[var][0,:,:]

            ax_0x = plt.subplot(3, 4, scen+1, projection=ccrs.PlateCarree())
            inarr0x.plot(ax = ax_0x, 
                       cmap = cmap,
                       cbar_kwargs   = {"label"       : "",
                                        "orientation" : "horizontal",
                                        "pad"         : 0})
            ax_0x.set_title(scenario)
            ax_0x.coastlines(resolution = 'auto',
                            linewidths =  0.75,
                            facecolor  = 'none')

    scen = 1
    if (inventory_row[scen+1]):
        scenario = "ssp245"
        
        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            inarr1x = xf[var][tind,:,:]

            ax_1x = plt.subplot(3, 4, scen+1, projection=ccrs.PlateCarree())
            inarr1x.plot(ax = ax_1x,
                       cmap = cmap,
                       cbar_kwargs   = {"label"       : "",
                                        "orientation" : "horizontal",
                                        "pad"         : 0})
            ax_1x.set_title(scenario)
            ax_1x.coastlines(resolution = 'auto',
                            linewidths =  0.75,
                            facecolor  = 'none')
        
    scen = 2      
    if (inventory_row[scen+1]):
        scenario = "ssp370"
        
        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            inarr2x = xf[var][tind,:,:]

            ax_2x = plt.subplot(3, 4, scen+1, projection=ccrs.PlateCarree())
            inarr2x.plot(ax = ax_2x,
                       cmap = cmap,
                       cbar_kwargs   = {"label"       : "",
                                        "orientation" : "horizontal",
                                        "pad"         : 0})
            ax_2x.set_title(scenario)
            ax_2x.coastlines(resolution = 'auto',
                            linewidths =  0.75,
                            facecolor  = 'none')
            
    scen = 3  
    if (inventory_row[scen+1]):
        scenario = "ssp585"
        
        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            inarr3x = xf[var][tind,:,:]

            ax_3x = plt.subplot(3, 4, scen+1, projection=ccrs.PlateCarree())
            inarr3x.plot(ax = ax_3x,
                       cmap = cmap,
                       cbar_kwargs   = {"label"       : "",
                                        "orientation" : "horizontal",
                                        "pad"         : 0})
            ax_3x.set_title(scenario)
            ax_3x.coastlines(resolution = 'auto',
                            linewidths =  0.75,
                            facecolor  = 'none')        


    
        
    
    # tasmin 
    var  = "tasmin"
    cmap = tmncmap
    
    scen = 0
    if (inventory_row[scen+1]):
        scenario = "historical"

        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            if (var in list(xf.keys())):
            
            
                inarr0n = xf[var][0,:,:]

                ax_0n = plt.subplot(3, 4, 4+scen+1, projection=ccrs.PlateCarree())
                inarr0n.plot(ax = ax_0n, 
                           cmap = cmap,
                           cbar_kwargs   = {"label"       : "",
                                            "orientation" : "horizontal",
                                            "pad"         : 0})
                ax_0n.set_title(scenario)
                ax_0n.coastlines(resolution = 'auto',
                                linewidths =  0.75,
                                facecolor  = 'none')

    scen = 1
    if (inventory_row[scen+1]):
        scenario = "ssp245"
        
        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            if (var in list(xf.keys())):
                inarr1n = xf[var][tind,:,:]

                ax_1n = plt.subplot(3, 4, 4+scen+1, projection=ccrs.PlateCarree())
                inarr1n.plot(ax = ax_1n,
                           cmap = cmap,
                           cbar_kwargs   = {"label"       : "",
                                            "orientation" : "horizontal",
                                            "pad"         : 0})
                ax_1n.set_title(scenario)
                ax_1n.coastlines(resolution = 'auto',
                                linewidths =  0.75,
                                facecolor  = 'none')
        
    scen = 2      
    if (inventory_row[scen+1]):
        scenario = "ssp370"
        
        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            if (var in list(xf.keys())):
                inarr2n = xf[var][tind,:,:]

                ax_2n = plt.subplot(3, 4, 4+scen+1, projection=ccrs.PlateCarree())
                inarr2n.plot(ax = ax_2n,
                           cmap = cmap,
                           cbar_kwargs   = {"label"       : "",
                                            "orientation" : "horizontal",
                                            "pad"         : 0})
                ax_2n.set_title(scenario)
                ax_2n.coastlines(resolution = 'auto',
                                linewidths =  0.75,
                                facecolor  = 'none')
        scen = 3  
    if (inventory_row[scen+1]):
        scenario = "ssp585"
        
        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            if (var in list(xf.keys())):
                inarr3n = xf[var][tind,:,:]

                ax_3n = plt.subplot(3, 4, 4+scen+1, projection=ccrs.PlateCarree())
                inarr3n.plot(ax = ax_3n,
                           cmap = cmap,
                           cbar_kwargs   = {"label"       : "",
                                            "orientation" : "horizontal",
                                            "pad"         : 0})
                ax_3n.set_title(scenario)
                ax_3n.coastlines(resolution = 'auto',
                                linewidths =  0.75,
                                facecolor  = 'none')  
    
    # prec 
    var = "pr"
    cmap = precmap
    
    scen = 0
    if (inventory_row[scen+1]):
        scenario = "historical"

        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            inarr0p = xf[var][0,:,:]

            ax_0p = plt.subplot(3, 4, 8+scen+1, projection=ccrs.PlateCarree())
            inarr0p.plot(ax = ax_0p, 
                       cmap = cmap,levels=clevs,
                       cbar_kwargs   = {"label"       : "",
                                        "orientation" : "horizontal",
                                        "pad"         : 0})
            ax_0p.set_title(scenario)
            ax_0p.coastlines(resolution = 'auto',
                            linewidths =  0.75,
                            facecolor  = 'none')

    scen = 1
    if (inventory_row[scen+1]):
        scenario = "ssp245"
        
        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            inarr1p = xf[var][tind,:,:]

            ax_1p = plt.subplot(3, 4, 8+scen+1, projection=ccrs.PlateCarree())
            inarr1p.plot(ax = ax_1p,
                       cmap = cmap,levels=clevs,
                       cbar_kwargs   = {"label"       : "",
                                        "orientation" : "horizontal",
                                        "pad"         : 0})
            ax_1p.set_title(scenario)
            ax_1p.coastlines(resolution = 'auto',
                            linewidths =  0.75,
                            facecolor  = 'none')
        
    scen = 2      
    if (inventory_row[scen+1]):
        scenario = "ssp370"
        
        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            inarr2p = xf[var][tind,:,:]

            ax_2p = plt.subplot(3, 4, 8+scen+1, projection=ccrs.PlateCarree())
            inarr2p.plot(ax = ax_2p,
                       cmap = cmap,levels=clevs,
                       cbar_kwargs   = {"label"       : "",
                                        "orientation" : "horizontal",
                                        "pad"         : 0})
            ax_2p.set_title(scenario)
            ax_2p.coastlines(resolution = 'auto',
                            linewidths =  0.75,
                            facecolor  = 'none')
    scen = 3  
    if (inventory_row[scen+1]):
        scenario = "ssp585"
        
        url    = root_url + scenario + "/LOCA2-CONUS___" + inventory_row.Model+"."+inventory_row.Member + "___" + scenario + ".nc"        
        if os.path.exists(url):
            print(url)
            xf     = xr.open_dataset(filename_or_obj = url)
            inarr3p = xf[var][tind,:,:]

            ax_3p = plt.subplot(3, 4, 8+scen+1, projection=ccrs.PlateCarree())
            inarr3p.plot(ax = ax_3p,
                       cmap = cmap,levels=clevs,
                       cbar_kwargs   = {"label"       : "",
                                        "orientation" : "horizontal",
                                        "pad"         : 0})
            ax_3p.set_title(scenario)
            ax_3p.coastlines(resolution = 'auto',
                            linewidths =  0.75,
                            facecolor  = 'none')  
        
    plt.suptitle(inventory_row.Model+"."+inventory_row.Member)    
    plt.tight_layout()
    plt.show()
    plt.clf()
    
    print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX")
```
