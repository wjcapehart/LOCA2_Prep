```python
import os    as os

import numpy as np

import xarray as xr

import pandas as pd

```

```python
def days_per_year(year):
    if (year % 4) == 0:
        if (year % 100) == 0:
            if (year %  400) == 0:
                days_per_year = 366
            else:
                days_per_year = 365
        else:
            days_per_year = 366
    else:
        days_per_year = 365
    return days_per_year
```

```python
working_directory  = "./"

loca2_inv_file     = "./LOCA2_Model_Member_Complete_List.csv"
loca2_inv_file     = "./LOCA2_Model_Member_Available_List.csv"

#loca2_repair_tasmin_file = "./FIX_TASMIN.csh"

loca2_remote_website = "https://cirrus.ucsd.edu/~pierce/LOCA2/NAmer/"

globus_root_dir    = "/data/GLOBUS/wjc"

thredds_root_dir   = "/data/DATASETS/LOCA_MACA_Ensembles/LOCA2/LOCA2_CONUS/Original_CONUS"


variables          = ["pr"]
var                = "pr"


scenarios          = ["historical", 
                      "ssp245", 
                      "ssp370", 
                      "ssp585"]


hist_year_strings  = ["1950-2014"] 


futr_year_strings  = ["2015-2044",
                      "2045-2074",
                      "2075-2100"]


version_strings    = ["LOCA_16thdeg_v20240915"] # pr
version_string     = version_strings[0]


bounds_script_command = 'defdim("bnds",2);lon_bnds=make_bounds(lon,$bnds,"lon_bnds");lat_bnds=make_bounds(lat,$bnds,"lat_bnds");time_bnds=make_bounds(time,$bnds,"time_bnds");'
```

```python
loca2_ensembles_list = pd.read_csv(filepath_or_buffer = loca2_inv_file)
models  = loca2_ensembles_list[ "Model"].values
members = loca2_ensembles_list["Member"].values

display(loca2_ensembles_list)

```

```python
mind     = 0
mind_end = -1

models  =  models[mind:mind_end]
members = members[mind:mind_end]

for m in range(len(models)):
    model  =  models[m]
    member = members[m]
    print( model + "." + member)

```

```python
display(loca2_ensembles_list.sort_values(by = "Model"))
```

```python
#for m in range(len(models)):
for m in [0]:
    
    model  = models[m]
    member = members[m]

    meta_deleteme = False


   
    meta_script_name = "./" + model + "." + member  + ".csh"
    print( model + "." + member)
    


    f0 = open(file     = meta_script_name, 
             mode      = 'w', 
             encoding  = "utf-8")

        



    
    
    for scenario in scenarios:
        
        deleteme = False
        
        
        script_name = "./" + model + "." + member + "." + scenario + ".csh"
        log_name    = "./" + model + "." + member + "." + scenario + ".out"

        f = open(file     = script_name, 
                 mode     = 'w', 
                 encoding = "utf-8")

        
        print("# XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX", file=f)
      
        if (scenario == "historical"):
            year_strings = hist_year_strings
        else:
            year_strings = futr_year_strings
            
        thredds_dir = thredds_root_dir + "/" + scenario

        fnl_thredds_file = "LOCA2-CONUS___" + model + "." + member + "___" + scenario + ".nc"
        fnl_thredds_loc  = thredds_dir + "/" + fnl_thredds_file

        
        pr_thredds_file = "LOCA2-CONUS___pr___" + model + "." + member + "___" + scenario + ".nc"
        pr_thredds_loc  = thredds_dir + "/" + pr_thredds_file

        xn_thredds_file = "LOCA2-CONUS___temps___" + model + "." + member + "___" + scenario + ".nc"
        xn_thredds_loc  = thredds_dir + "/" + xn_thredds_file
        
        cat_thredds_pr_yrs_file =  "LOCA2-CONUS___pr___"  + model + "." + member + "___" + scenario + "___Y-*.nc"
        cat_thredds_pr_yrs_loc  = thredds_dir + "/" + cat_thredds_pr_yrs_file

        xnp_thredds_file = "LOCA2-CONUS___xnp___" + model + "." + member + "___" + scenario + ".nc"
        xnp_thredds_loc  = thredds_dir + "/" + xnp_thredds_file

        print('nohup cdo --no_history  -f nc4 -z zip_8  delname,pr '+  fnl_thredds_loc + ' ' + xn_thredds_loc + " &", file=f)

        
        for year_string in year_strings:
            i_fixed_fixme = True

            print("# =========================================", file=f)
            year_bounds = year_string.split(sep="-")
            years_num   = np.arange(int(year_bounds[0]),int(year_bounds[1])+1)
            years_str   = str(years_num).zfill(4)
            year0       = 0
            yearN       = 0


            globus_file = "pr." + model + "." + scenario + "." + member + "." + year_string + "." + version_string + ".nc"

            #
            # Pull precip file from Dave's website
            #


            url_ucsd    = loca2_remote_website + model + "/0p0625deg/" + member + "/" + scenario  + "/"+ var + "/" + globus_file

            globus_dir  = "./"         

            globus_file_loc  = globus_dir + globus_file

            print('nohup wget  --no-verbose --output-document=' + globus_file_loc + '  ' + url_ucsd, file=f)
            print(" ", file=f)



            #
            # Clean NetCDF Metadata
            #

            print('nohup ncatted -h -O -a           ,global,d,,             ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a           history,,d,,            ' + globus_file_loc, file=f)

            print(" ", file=f)

            print('nohup ncatted -h -O -a              ,time,d,,             ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a   description,time,c,c,"time"      ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a     long_name,time,c,c,"time"      ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a standard_name,time,c,c,"time"      ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a         units,time,c,c,"days since 1900-01-01" ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a      calendar,time,c,c,"standard"  ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a          axis,time,c,c,"T"         ' + globus_file_loc, file=f)

            print(" ", file=f)

            print('nohup ncatted -h -O -a              ,lat,d,,                 ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a   description,lat,c,c,"Latitude"      ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a     long_name,lat,c,c,"Latitude"      ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a standard_name,lat,c,c,"latitude"      ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a         units,lat,c,c,"degrees_north" ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a          axis,lat,c,c,"Y"             ' + globus_file_loc, file=f)

            print(" ", file=f)

            print('nohup ncatted -h -O -a              ,lon,d,,                ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a   description,lon,c,c,"Longitude"    ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a     long_name,lon,c,c,"Longitude"    ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a standard_name,lon,c,c,"longitude"    ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a         units,lon,c,c,"degrees_east" ' + globus_file_loc, file=f)
            print('nohup ncatted -h -O -a          axis,lon,c,c,"X"            ' + globus_file_loc, file=f)

                
            for year in years_num:
                print("# -----------------------------------------", file=f)

                single_year_string = "Y-" + str(year).zfill(4)

                dpy   = days_per_year(year)
                yearN = year0 + dpy - 1

                mid_wildthredds_file = "LOCA2-CONUS___pr___" + model + "." + member + "___" + scenario + "___" + single_year_string + ".nc"
                cat_thredds_file     = "LOCA2-CONUS___pr___" + model + "." + member + "___" + scenario + "___" + single_year_string + ".nc"

                mid_wildthredds_loc = thredds_dir + "/" + mid_wildthredds_file
                cat_thredds_loc     = thredds_dir + "/" + cat_thredds_file



                for v in range(len(variables)):

                    variable       =       variables[v]
                    version_string = version_strings[v]
                    
                    

                    mid_thredds_file     = "LOCA2-CONUS___" + variable + "___" + model + "." + member + "___" + scenario + "___" + single_year_string + ".nc"
                    mid_thredds_loc      = thredds_dir + "/" + mid_thredds_file

                    print(" ", file=f)  

                    #
                    # Extract Year
                    #

                    print('nohup ncks -O -4 -h -d time,' + str(year0) + ',' + str(yearN) + ' ' + globus_file_loc + ' ' + mid_thredds_loc, file=f)

                    print(" ", file=f)

                    #
                    # Correct Coordinate Variable Types
                    #

                    print('nohup ' + "ncap2 -O -4 -h -s 'time = double(time);lat = float(lat);lon = float(lon);' " + mid_thredds_loc + ' ' + mid_thredds_loc + '.temp.nc', file=f)
                    print('nohup mv -fv ' + mid_thredds_loc + '.temp.nc' + ' ' +  mid_thredds_loc   , file=f)

                    print(" ", file=f)

                    #
                    # Compress Variable to short integer
                    #                    

                    compress_cmd = "where(pr > 1e9) pr=short(round(-32767)); elsewhere pr=short(round(pr*864000.));"

                    print('nohup ncatted -h -O -a              ,pr,d,,                             ' + mid_thredds_loc, file=f)

                    print("nohup ncap2 -O -4 -h --script '" + compress_cmd + "' " + mid_thredds_loc + ' ' + mid_thredds_loc + '.temp.nc' , file=f)
                    print('nohup mv -fv ' + mid_thredds_loc + '.temp.nc' + ' ' +  mid_thredds_loc   , file=f)
                    print('nohup ' + "ncap2 -O -4 -h --script 'pr=short(pr)' " + mid_thredds_loc + ' ' + mid_thredds_loc + '.temp.nc' , file=f)
                    print('nohup mv -fv ' + mid_thredds_loc + '.temp.nc' + ' ' +  mid_thredds_loc   , file=f)

                    print(" ", file=f)

                    #
                    # Enforce Uniform Mask
                    #

                    print('nohup cdo --no_history mul '+  mid_thredds_loc + ' ~/GitHub/LOCA2_Prep/loca2_mask.nc ' + mid_thredds_loc + '.temp.nc', file=f)
                    print('nohup mv -fv '  + mid_thredds_loc + '.temp.nc' + ' ' +  mid_thredds_loc   , file=f)

                    print(" ", file=f)

                    #
                    # Create Standardized Metadata
                    #

                    print('nohup ncatted -h -O -a   description,pr,c,c,"Daily Total Precipitation" ' + mid_thredds_loc, file=f)
                    print('nohup ncatted -h -O -a     long_name,pr,c,c,"Daily Total Precipitation" ' + mid_thredds_loc, file=f)
                    print('nohup ncatted -h -O -a standard_name,pr,c,c,"precipitation_amount"      ' + mid_thredds_loc, file=f)
                    print('nohup ncatted -h -O -a         units,pr,c,c,"kg m-2"                    ' + mid_thredds_loc, file=f)
                    print('nohup ncatted -h -O -a  cell_methods,pr,c,c,"time: sum"                 ' + mid_thredds_loc, file=f)
                    print('nohup ncatted -h -O -a    _FillValue,pr,c,s,-32767                      ' + mid_thredds_loc, file=f)
                    print('nohup ncatted -h -O -a missing_value,pr,c,s,-32767                      ' + mid_thredds_loc, file=f)
                    print('nohup ncatted -h -O -a    add_offset,pr,c,f,0                           ' + mid_thredds_loc, file=f)
                    print('nohup ncatted -h -O -a  scale_factor,pr,c,f,0.1                         ' + mid_thredds_loc, file=f)

                    print(" ", file=f)

                    year0 = yearN+1

                    

        #
        # Merge all years
        #
        print("# -----------------------------------------", file=f)

        print('nohup cdo  --no_history -f nc4 -z zip_8  mergetime '+ cat_thredds_pr_yrs_loc + ' ' + pr_thredds_loc, file=f)

        #
        # Apply Consistent Time Bounds
        #

        print(" ", file=f)
        print('nohup ' + "ncap2 -O -4 -h -s '" + bounds_script_command + "' " + pr_thredds_loc + ' ' + pr_thredds_loc + '.temp.nc', file=f)
        print('nohup mv -fv ' + pr_thredds_loc + '.temp.nc' + ' ' +  pr_thredds_loc   , file=f)
        print(" ", file=f)

        #
        # Apply Global Metadata
        #


        print('nohup ncatted -h -O -a  title,global,c,c,"LOCA statistically downscaled climate model data, David W. Pierce, Scripps Institution of Oceanography" ' + pr_thredds_loc, file=f)
        print('nohup ncatted -h -O -a  author,global,c,c,"David W. Pierce, Scripps Institution of Oceanography" ' + pr_thredds_loc, file=f)
        print('nohup ncatted -h -O -a  cmip_id,global,c,c,"CMIP6" '                         + pr_thredds_loc, file=f)
        print('nohup ncatted -h -O -a  experiment_id,global,c,c,"'    + scenario + '" '     + pr_thredds_loc, file=f)
        print('nohup ncatted -h -O -a  model_source_id,global,c,c,"'  + model    + '" '     + pr_thredds_loc, file=f)
        print('nohup ncatted -h -O -a  model_variant_id,global,c,c,"' + member   + '" '     + pr_thredds_loc, file=f)
        print('nohup ncatted -h -O -a  Conventions,global,c,c,"CF-1.7 CMIP-6.0 UGRID-1.0" ' + pr_thredds_loc, file=f)
        
        print("rm -frv "+cat_thredds_pr_yrs_loc, file=f)
        print("rm -frv "+globus_file_loc,     file=f)
            
        print("# =========================================", file=f)

        print('nohup cdo  --no_history -f nc4 -z zip_8 merge '+ xn_thredds_loc + ' ' + pr_thredds_loc + " ./temp.nc", file=f)

        
        print("# XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX", file=f)

        print("# -----------------------------------------", file=f)

        

        f.close()
        os.system("chmod a+x " + script_name)  

        f.close()
        if deleteme:
            os.system("rm " + script_name)
        else:    
            print("nohup " + script_name + " &> " + log_name,file=f0)
            os.system("chmod a+x " + script_name)  

    f0.close()
    if meta_deleteme:
        os.system("rm " + meta_script_name)
    else:
        os.system("chmod a+x " + meta_script_name)   

    os.system("chmod a+x " + meta_script_name)            

        
```
