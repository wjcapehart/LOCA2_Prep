```python
import xarray as xr
import numpy as np
import pandas as pd
import cartopy.crs       as ccrs
import cartopy.feature   as creature
import netCDF4           as ncf
import matplotlib.pyplot as plt


```

```python
ds_ms = xr.open_dataset("./LOCA2_MASKS.nc")

ds_ll = xr.open_dataset("./latlons.nc")
```

```python
plt.pcolormesh(ds_ms["LOCA2_MASK"])
```

```python
ds_mask = ds_ll.copy()
maskarr = ds_ms["LOCA2_MASK"].values.astype(np.int16)
maskarr[maskarr < 1] = np.int16(-32767),
ds_mask["loca2_mask"] = xr.DataArray(data=maskarr,
                                     coords=[ds_mask["lat"],
                                           ds_mask["lon"]],
                                     attrs={"_FillValue": np.int16(-32767),
                                            "description":"mask",
                                            "long_name":"mask",
                                            "standard_name":"land_binary_mask",
                                            "units":"1"})

del ds_mask["pr"]
del ds_mask["time"]
display(ds_mask)
print(np.nanmax(ds_mask["loca2_mask"]))
ds_mask["loca2_mask"].plot()
ds_mask.to_netcdf("./loca2_mask.nc")
```
