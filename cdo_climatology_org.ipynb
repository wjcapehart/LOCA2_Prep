```python
def days_per_year(year):
    if (year % 4) == 0:
        if (year % 100) == 0:
            if (year %  400) == 0:
                days_per_year = 366
            else:
                days_per_year = 365
        else:
            days_per_year = 366
    else:
        days_per_year = 365
    return days_per_year
```
